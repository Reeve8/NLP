#Regular Expression
##RegEx, or Regular Expression, is a sequence of characters that forms a search pattern. RegEx can be used to check if a string contains the specified search pattern.


#RegEx Module
##Python has a built-in package called re, which can be used to work with Regular Expressions.

#Import the re module:

```python
import re
```

```python
txt = "It is a Beautiful day"
x = re.search("^It.*day$", txt)
x
```

```python
import re
txt = "Wow Wonderful day"
x = re.search(".*day$", txt)
if x:
  print("YES! We have a match!")
else:
  print("No match")
```

```python
import re
txt = "Wow Wonderful day"
x = re.search("Wo.*day$", txt)
if x:
  print("YES! We have a match!")
else:
  print("No match")
```

#RegEx Functions
##The re module offers a set of functions that allows us to search a string for a match:


##findall	- Returns a list containing all matches
##search	- Returns a Match object if there is a match anywhere in the string
##split	- Returns a list where the string has been split at each match
##sub	- Replaces one or many matches with a string

```python
txt = "Wow! it is a Wonderful day"
x = re.search("Wo", txt)
print(x)
```

```python
txt = "It is a Beautiful day"
x = re.split("\s", txt)
print(x)
```

```python
#Replace all white-space characters with the digit "_":
import re
txt = "It is a Beautiful day"

x = re.sub("\s", "-", txt)
print(x)
```

##span() returns a tuple containing the start-, and end positions of the match.
##string returns the string passed into the function
##group() returns the part of the string where there was a match

```python
txt = "Wow! it is a Wonderful day"
x = re.search(r"\bW\w+", txt)
print(x.span())
x
```

```python
x = re.search(r"\bd\w+", txt)
print(x.span())
```

```python
x = re.search(r"\bd\w+", txt)
print(x.string)
```

```python
import re
txt = "Wow! it is a Wonderful day"
x = re.match(r"(\bW\w+)", txt)
print(x.group())
x
```

```python
txt = "Wow! it is a Wonderful day"
x = re.search(r"\bW\w+", txt)
print(x.group())
```

The \r metacharacter matches carriage return characters.
The carriage return in Python is defined with the escape character \r. This escape character allows you to reset the position of your cursor to the beginning of a line of text. When you use this escape character, the text at the beginning will be overwritten with the text after the \r character.

```python
#A set of characters
txt = "this toy costs 450 rupees"
x = re.findall("[c-i]", txt)
print(x)
```

```python
#Signals a special sequence (can also be used to escape special characters)
x = re.findall("\d", txt)
print(x)
```

```python
#Any character (except newline character)
x = re.findall("r....s", txt)
print(x)
```

```python
#starts with
x = re.findall("^toy", txt)
if x:
  print("Yes, the string starts with 'this'")
else:
  print("No match")
```

```python
#starts with
x = re.findall("toy$", txt)
if x:
  print("Yes, the string starts with 'this'")
else:
  print("No match")
```

```python
#starts with
x = re.findall("rupees$", txt)
if x:
  print("Yes, the string starts with 'this'")
else:
  print("No match")
```

```python
txt = "this toy costs 450 rupees"
#	Zero or more occurrences
x = re.findall("ru.*s", txt)
print(x)
```

```python
#One or more occurrences
x = re.findall("ru.+s", txt)
print(x)
```

```python
x = re.findall("rup.?s", txt)
print(x)
##This time we got no match, because there were not zero, not one, but two characters between "ru" and the "e"
```

```python
x = re.findall("to.?y", txt)
print(x)
```

```python
#Exactly the specified number of occurrences

x = re.findall("rup.{2}s", txt)

print(x)
```

```python
x = re.findall("toy|costs", txt)

print(x)

if x:
  print("Yes, there is at least one match!")
else:
  print("No match")
```

```python
x = re.findall("it|pay", txt)

print(x)

if x:
  print("Yes, there is at least one match!")
else:
  print("No match")
```

```python
x = re.findall("toy|pay", txt)

print(x)

if x:
  print("Yes, there is at least one match!")
else:
  print("No match")
```
